```python
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',on_bad_lines='skip',encoding='latin-1')
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user-id','Book-Rating':'book-rating'},inplace=True)
```

```python
books.shape
```

```python
x=ratings['user-id'].value_counts()>200
```

```python
y=x[x].index
```

```python
x[x].shape
```

```python
y
```

```python
ratings=ratings[ratings['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['book-rating'].count().reset_index()
```

```python
number_rating.rename(columns={'book-rating':'number of ratings'}, inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user-id',index='title',values='book-rating')
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions=model.kneighbors(book_pivot.iloc[240,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distances, suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
  for i in range(len(suggestions)):
    if i !=0:
      print('The suggestions for',book_name,"are:")
    else:
      print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
